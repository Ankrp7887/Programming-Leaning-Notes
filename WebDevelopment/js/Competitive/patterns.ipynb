```java
int rows = 2, columns = 2;

for(int i=0; i<rows; i++) {
    for(int j=0;j<columns;j++) 
        System.out.print("*");
    System.out.println();
}

System.out.println();

for(int i=1; i<=rows; i++) {
    for(int j=1;j<=columns;j++) 
        System.out.print(i);
    System.out.println();
}

System.out.println();

for(int i=rows; i>0; i--) {
    for(int j=columns; j>0; j--) 
        System.out.print(i);
    System.out.println();
}

System.out.println();

for(int i=1; i<=rows; i++) {
    for(int j=1;j<=columns;j++) 
    System.out.print(j);
    System.out.println();
}

System.out.println();

for(int i=rows; i>0; i--) {
    for(int j=columns; j>0; j--) 
        System.out.print(j);
    System.out.println();
}

System.out.println();

int num = 1;
for(int i=1; i<=rows; i++) {
    for(int j=1;j<=columns;j++){ 
        System.out.print(num++ + " ");
    }
    System.out.println();
}

System.out.println();

int num = rows*columns;
for(int i=1; i<=rows; i++) {
    for(int j=1;j<=columns;j++){ 
        System.out.print(num-- + " ");
    }
    System.out.println();
}
```

```java
int rows = 3;

for(int i=0; i<rows; i++) {
    for(int k=rows-1; k>i ;k--)
        System.out.print(" ");
    for(int j=0;j<=i;j++) 
        System.out.print("*");
    System.out.println();
}

System.out.println();

for(int i=1; i<=rows; i++) {
    for(int k=rows-1; k>=i ;k--)
        System.out.print(" ");
    for(int j=1;j<=i;j++) 
        System.out.print(i);
    System.out.println();
}

System.out.println();

for(int i=rows; i>0; i--) {
    for(int k=1;k>(2-i);k--)
        System.out.print(" ");
    for(int j=rows;j>=i;j--) 
        System.out.print(i);
    System.out.println();
}

System.out.println();

for(int i=1; i<=rows; i++) {
    for(int k=rows-1; k>=i ;k--)
        System.out.print(" ");
    for(int j=1;j<=i;j++) 
        System.out.print(j);
    System.out.println();
}

System.out.println();

for(int i=rows; i>0; i--) {
    for(int k=1;k>(2-i);k--)
        System.out.print(" ");
    for(int j=rows;j>=i;j--) 
        System.out.print(j);
    System.out.println();
}

System.out.println();

int num = 1;
for(int i=1; i<=rows; i++) {
    for(int k=rows-1; k>=i ;k--)
        System.out.print(" ");
    for(int j=1;j<=i;j++) 
        System.out.print(num++ + " ");
    System.out.println();
}

System.out.println();

int num = (rows*(rows+1))/2; // (rows*(rows+1))/2
for(int i=1; i<=rows; i++) {
    for(int k=rows-1; k>=i ;k--)
        System.out.print(" ");
    for(int j=1;j<=i;j++) 
        System.out.print(num-- + " ");
    System.out.println();
}
```
